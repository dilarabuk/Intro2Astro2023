We can sort data according to a column values.

-ascending = by ascending order
-inplace = Change the array itself

```python
import pandas as pd
data = pd.DataFrame({'group':['a', 'a', 'a', 'b','b', 'b', 'c', 'c','c'],'ounces':[4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data.sort_values(by=['ounces'],ascending=True,inplace=True)
data
```

We need to tie the food and animal column now. In other words food is the "Foreign Key" here.

-Create another dictionary now. (Key is foodtypes and value is animaltypes.)

Now need to explain .map. 
-.map allows you to apply a function to each item in an iterable (such as a list, tuple, or dictionary) and return a new iterable (a map object) that contains the results of the function.

to find each element of data['animal'] ; 
-check all data['food']
-Turn their all letters into lowercase
-And give the results of from meat_to_animal by using associated data['food']

```python
import pandas as pd
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami','corned beef', 'Bacon', 'pastrami', 'honey ham','nova lox'],
                 'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})

meat_to_animal = {
'bacon': 'pig',
'pulled pork': 'pig',
'pastrami': 'cow',
'corned beef': 'cow',
'honey ham': 'pig',
'nova lox': 'salmon'
}

data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```
